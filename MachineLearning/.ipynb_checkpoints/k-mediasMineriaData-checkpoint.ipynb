Algoritmo de k-medias
El algoritmo de las k-medias es un algoritmo de aprendizaje no supervisado para
resolver el problema de la clusterización.
Tiene como objetivo la partición de un conjunto de "n" observaciones en "k" grupos,
en el que cada observación pertenece al grupo cuyo valor medio es más cercano.
Es un método utilizado en la minería de datos.

```python
import pandas as pd
import numpy as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
datos = make_blobs(n_samples= 200, n_features =2, centers=4 )
datos
```

```python
datos[1]
```

```python
plt.scatter(datos[0][:,0], datos[0][:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters= 4)
```

```python
modelo.fit(datos[0])
```

```python
modelo.cluster_centers_ #los puntos centrales de cada grupo
```

```python
modelo.labels_
```

```python
datos[1]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.scatter(datos[0][:,0], datos[0][:,1], c=modelo.labels_)
ax1.set_title('Algoritmo de k-medias')
ax2.scatter(datos[0][:,0], datos[0][:,1], c=datos[1])
ax2.set_title('Datos originales')
```
