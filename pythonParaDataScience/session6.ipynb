```python
sns.set_style("dark", {"axes.facecolor":"black"})#white, whitegrid, dark, darkgrid, ticks
f, axes = plt.subplots(2,2, figsize=(15,15))
#Grafico [0,0]
k1 = sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsAudiencia,\
                 shade=True, shade_lowest=True, cmap='inferno', ax=axes[0,0])
k1b = sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsAudiencia,\
                 cmap='cool', ax=axes[0,0])
#Grafico [0,1]
k2 = sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsCriticos,\
                 shade= True, shade_lowest=True, cmap='inferno', ax=axes[0,1])
k2b = sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsCriticos,\
                 cmap='cool', ax=axes[0,1])
#Grafico [1,0]
z = sns.violinplot(data=peliculas, x='Año', y='PresupuestoMillones', ax=axes[1,0], palette='YlOrRd')
#Grafico [1,1]

k4= sns.kdeplot(x=peliculas.RatingsCriticos, y=peliculas.RatingsAudiencia,\
                shade=True, shade_lowest=False, cmap='Blues_r', ax=axes[1,1])

k4b= sns.kdeplot(x=peliculas.RatingsCriticos, y=peliculas.RatingsAudiencia,\
                cmap='gist_gray_r', ax=axes[1,1])
# axes[1,1].hist(peliculas.RatingsAudiencia, bins = 15)
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
#Cambios Temáticos
```

```python
sns.set_style("whitegrid")
lista1 = list()
mis_etiquetas = list()
for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == gen].PresupuestoMillones)
    mis_etiquetas.append(gen)

fig, ax = plt.subplots(1,1)
fig.set_size_inches(11.7, 8.27)
h = plt.hist(lista1, bins=30, stacked=True, rwidth=1, edgecolor='none', label=mis_etiquetas )
plt.title('Distribución de Presupuestos para las Películas', fontsize=35, color="DarkBlue" )#fontname="Console"
plt.ylabel("Número de películas", fontsize=25, color="Red")
plt.xlabel("Presupuesto", fontsize=25, color="Green")
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(frameon=True, fancybox=True, framealpha=1, shadow=True, prop={'size':20})
plt.show()
#Para profundizar puedo ponerme a cambiar colores del hist, habría que pasarle un diccionario
```
